### raw data

```python


# read data
raw_1 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data1.csv')
raw_2 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data2.csv')
raw_3 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data3.csv')
raw_4 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data4.csv')
raw_5 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data5.csv')
raw_6 = pd.read_csv('https://api.covid19india.org/csv/latest/raw_data6.csv')

# concatenate dfs
full_data = pd.concat([raw_1, raw_2, raw_3, raw_4, raw_5, raw_6])

# save to csv`
full_data.to_csv('data/patients_data.csv', index=False)

# first few rows
full_data.head()

```

### state_wise_daily

```python
# read data
state_wise_daily = pd.read_csv('https://api.covid19india.org/csv/latest/state_wise_daily.csv')
# change wider data into longer
state_wise_daily = state_wise_daily.melt(id_vars = ['Date', 'Status'])

#pivot operation
state_wise_daily = state_wise_daily.pivot_table(values = 'value', index = ['Date', 'variable'], columns='Status').reset_index()
state_wise_daily.rename(columns = {'variable':'State'}, inplace=True)

# State Name mapping
state_map = {state:state_name for state, state_name in zip(state_level_latest['State_code'], state_level_latest['State'])}
state_map['DD']='Daman & Diu'
state_wise_daily['State_name'] = state_wise_daily['State'].map(state_map)

# Saving to csv file
state_wise_daily.to_csv('data/state_level_daily.csv')

# first few rows
state_wise_daily.head()
```
